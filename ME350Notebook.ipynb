```python
# The first step for any Python code is to Import Libraries
import numpy as np               # numpy is the library for many useful math functions
from scipy import integrate      # here, we specifically import a single function, 'integrate'
import matplotlib.pyplot as plt  # for plotting. Note, "as" creates a shortcut to save typing time

# Input constants
delta_t = 0.02                   # time step
t_max = 10                       # time to end simulation
theta1_0 = np.pi/2               # initial displacement
theta2_0 = 0                     # initial angular velocity
theta_init = (theta1_0, theta2_0)
# Get timesteps
t = np.linspace(0, t_max, int(t_max/delta_t))

```

Recall that $\ddot{\theta}+\frac{c}{m}\dot{x}+\frac{g}{l}x=0 $

```python
def int_pendulum_sim(theta_init, t):
    m = 1                        # mass in kg
    L = 1                        # length in m
    b = 0.6                      # damping constant
    g = 9.81                     # gravitational constant
    theta_dot_1 = theta_init[1]
    theta_dot_2 = -b/m*theta_init[1] - g/L*np.sin(theta_init[0])
    return theta_dot_1, theta_dot_2

theta_vals_int = integrate.odeint(int_pendulum_sim, theta_init, t)

plt.plot(t, theta_vals_int[:,0], label='Position')
#plt.plot(t, theta_vals_int[:,1], label='Velocity')
plt.xlim(0,10)

#plt.plot(x, y1, 'bo-', linewidth=2, markersize=12, label='First');

plt.xlabel('$t$ [s]')
plt.ylabel('Angular position [rad]')
#plt.axis([-0.5, 10.5, -5, 105])
plt.legend(loc='upper right')
#plt.savefig("myplot.pdf")

#plt.plot(t, theta_vals_int[:,0])
plt.show()
```
